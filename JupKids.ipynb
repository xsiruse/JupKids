```python
import pandas as pd
```

Функция, которая загружает указанные года

```python
def load_years(years_list):
    names_by_year = {}
    for year in years_list:
        names_by_year[year] = pd.read_csv(
            f'names/yob{year}.txt',
            names=['Name','Gender','Count']
        )
    names_all = pd.concat(names_by_year).reset_index(level=1, drop=True)
    names_all = names_all.rename_axis(index='Year', columns="")
    return names_all
    
```

Пробуем на списке

```python
load_years([1900, 1950, 2000])
```

пробуем на рендже

```python
load_years(list(range(1880, 2017)))
```

Функция, которая выводит ТОП-Х

```python
def show_top_for_years(top, years):
    return load_years(years).groupby(['Name','Gender']).sum().sort_values('Count', ascending=False)[0:top]   
```

```python
show_top_for_years(3, [1880])
```

```python
show_top_for_years(3, [1900, 1950, 2000])
```

```python
show_top_for_years(10, list(range(1880, 2017)))
```

Функция, которая возвращает динамику изменения количества имен за указанные года в разрезе полов.

```python
def dynamics_gender(years):
    dynam = {}
    for year in years:
        df = load_years([year])
        dynam.setdefault('M', []).append(df[df['Gender'] == 'M']['Count'].sum())
        dynam.setdefault('F', []).append(df[df['Gender'] == 'F']['Count'].sum())
    return dynam
```

```python
dynamics_gender([1900, 1950, 2000])
```
